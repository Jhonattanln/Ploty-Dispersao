## Criando gráfico de dispersão com Plotly

```python
#### Importanto bibliotecas
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
import numpy as np
from plotly.subplots import make_subplots
import plotly.graph_objects as go
```

#### Tratando dados da série histórica de preços

```python
cot = pd.read_excel('economatica_fechamento.xlsx',index_col=0, skiprows=3, parse_dates=True)
```

```python
def columns(df): #renomeando colunas com as ações
    df.columns = df.columns.str[39:]
columns(cot)
cot.head()
```

```python
cot.replace('-', np.nan, inplace=True) ### removendo datas ausentes
cot.isna().sum() ### checando a quantidade de dados ausentes
```

```python
cot_2021 = cot.loc['2021'] ### selecionando o ano de analise 
cot_2021.isna().sum().sort_values(ascending=False).head(20) ### contagem de valores nulos
```

```python
cot_2021.drop(['CMIN3', 'ASAI3', 'IGTI11'], axis=1, inplace=True) ### excluindo colunas com valores faltantes
cot_2021 = cot_2021.dropna()
cot_2021_pct = cot_2021.pct_change().dropna() ### calculando retornos
```

```python
cot_2021_pct.describe()
```

#### Calculando a volatilidade

```python
### Calculo da volatilidade por janela
vol_2021 = pd.DataFrame()

for i in cot_2021_pct.columns:
    window = 21 ### selecionando janelas de 21 dias
    vol_2021[i] = cot_2021_pct[i].rolling(window).std()*(252**0.5)*100 
    vol_2021.dropna()
```

```python
### Botões
datas = [{'count':1, 'label':'1Mêses', 'step':'month', 'stepmode':'backward'},
{'count':6, 'label':'6Mêses', 'step':'month', 'stepmode':'backward'},
{'count':1, 'label':'YTD', 'step':'year', 'stepmode':'todate'}]
```

```python
fig = px.line(
    data_frame=vol_2021,
    title='Volatilidade 21 dias',
    x=vol_2021.index,
    y=vol_2021.columns)

fig.update_layout(
  	{'xaxis':
    {'rangeselector': {'buttons': datas}}})

fig.show()
```

```python
vol_2021.describe()
```

```python
#### Calculo da volatilidade anuas
vol = []
vol = np.std(cot_2021_pct)*252**(0.5)*100
vol = pd.DataFrame(vol, columns=['Volatilidade'])
```

#### Calculando rentabilidade

```python
ret = {}
for i in cot_2021.columns:
    ret[i] = (cot_2021[i].iloc[-1]/cot_2021[i].iloc[1]-1)*100
ret = pd.DataFrame(list(ret.items()), columns=['Ativo', 'Retorno'])
ret.set_index('Ativo', inplace=True)
ret.head()
```

```python
### Dados da participação dos ativos
dados = pd.read_excel('economatica_dados.xlsx', skiprows=3, index_col=6)
dados.head()
```

```python
### Concatenando dados
vol_part = pd.concat([vol, dados['Participação\nno Ibovespa\n\nem %'], ret, dados['Subsetor Bovespa']], axis=1).dropna()
vol_part.reset_index(inplace=True)
vol_part.rename(columns={'Participação\nno Ibovespa\n\nem %' : 'Participação no Ibovespa', 'index':'Ativo'}, inplace=True)
```

```python
vol_part.sort_values('Participação no Ibovespa', ascending=False)
```

```python
fig = px.scatter(
    data_frame=vol_part,
    x='Volatilidade',
    y='Retorno',
    size='Participação no Ibovespa',
    color='Subsetor Bovespa',
    hover_name='Ativo',
)
fig.update_layout({'title': {'text': 'Retorno x Volatilidade', 'x':0.5, 'y':0.95}})
fig.show()
```

```python
fig = make_subplots(
    rows=2, cols=1,
    subplot_titles=['Box retornos', 
    'Box volatilidade'])

fig.update_layout({'title': {'text': 'Estatística resultados', 'x':0.5, 'y':0.95}})

fig.add_trace(
    go.Box(x=vol_part.Retorno,
    name='Retornos',
    hovertext=vol_part.Ativo,
    showlegend=True),
    row=1, col=1)

fig.add_trace(
    go.Box(x=vol_part.Volatilidade,
    hovertext=vol_part.Ativo,
    name='Volatilidade'),
    row=2, col=1)

#fig = make_subplots(rows=2, cols=1, shared_xaxes=True)

fig.show()
```

```python
fig = go.Figure()
for i in vol_part['Subsetor Bovespa']:
    fig.add_trace(go.Bar(y=vol_part['Retorno'], x=vol_part['Ativo'], name=i))

buttons=[
    {'label': "Petróleo gás e biocombustíveis", 'method':"update", 'args':[{'visible':[True, False]},
    {'title': "Petróleo gás e biocombustíveis"}]},
    {'label': "Bebidas", 'method':"update", 'args':[{'visible':[False, True]}, 
    {'title': 'Bebidas'}]},
    ]

fig.update_layout({'updatemenus': [{
    'type':"dropdown",
    'showactive':True,
    'active':0, 
    'buttons': buttons}]})

fig.show()
```
